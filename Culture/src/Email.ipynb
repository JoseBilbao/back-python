```python
data = pd.read_csv('../data/hrdata_OSF.csv')

employee = {}
for emp in data.to_dict('records'):
    employee[emp['name']] = {
        "bank": emp['bank'],
        "acquired": emp['acquired'],
        "tenure": emp['tenure'],
        "title_status": emp['title_status'],
        "male": emp['male'],
        "rating": emp['rating'],
        "branch_id": emp['branch_id']
    }
```

```python
message = pd.read_csv('../data/SMS.csv')

messages = {}
for m in message.to_dict('records'):
    messages[m['sender']] = {
        'body':m['Message_body'],
        'receiver':m['receivers']
    }
    
```
